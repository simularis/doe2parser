```python
%config Completer.use_jedi = False
import graphviz
import ipywidgets as widgets
```

```python
graphviz.ENGINES
```

```python
def drawloop(**kwargs):
    g=graphviz.Digraph(**kwargs)
    r=range(4)
    for i in r:
        g.node(f"{i}")
    for i in r:
        g.edge(f"{i}",f"{r[i-1]}")
    return g
drawloop(engine="circo")
```

```python
for e in graphviz.ENGINES:
    print(e)
    display(drawloop(engine=e))
```

```python
g=drawloop(engine="circo")
g.node("4")
g.edge("4","0")
g.edge("0","4")
g.node("5")
g.edge("5","1")
g.edge("1","5")
g.node("6")
g.edge("6","0")
g.edge("2","6")
g
```
